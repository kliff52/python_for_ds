### Формирования текста для графики 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams 

%matplotlib inline 
rcParams['figure.figsize'] = 6,4.5
```

```python
x = np.linspace(-5,5,101) # Числовой ряд от -5 до 5 с шагом 0.1
x
```

```python
x2 = x**2 # Квадрат каждого значения
x2
```

```python
plt.plot(x, x2)
plt.title('График f(x)') #Название
plt.show
```

```python
# SVG (Scalable Vector Graphics) - маштабируемая векторная графика 
%config InlineBackend.figure_format = 'svg'
```

```python
#более четкий векторный график
plt.plot(x, x2)
plt.title('График f(x)', fontsize = 16, fontweight = 'bold', color = 'blue')  # Размер шрифта, насыщенность текста,
plt.show                                                                      # цвет по HTML названию или коду. 
```

Можно обьявить все отдельно. 

```python
title_dict= {'fontsize': 20, 'fontweight' : 'bold', 'color' : 'blue', 'family':'Calibri' }
plt.plot(x, x2)
plt.title('График f(x)', fontdict = title_dict, loc = 'left') # Располодения загаловка графика 
plt.show
```

добовление названия осей

```python
label_dict = {'fontsize':'22', 'color':'red', 'family':'Calibri'}
plt.plot(x, x2)
plt.title('График f(x)' , fontdict = title_dict) 
plt.xlabel('x', fontdict = label_dict) # Горизонт
plt.ylabel('f(x)', fontdict = label_dict) # Вертикаль
plt.show
```

Добавим график еще одной функции 

```python
x3 = x**3
x3
```

```python
plt.plot(x, x2)
plt.plot(x, x3)

plt.title('Граффики f(x)' , fontdict = title_dict) 
plt.xlabel('Переменная Х' , fontdict = label_dict) # Горизонт  
plt.ylabel('Функция f(x)' , fontdict = label_dict) # Вертикаль
plt.show
```

### Ограничения Осей.

```python
x4 = x**4
x4
```

```python
plt.axis([-5,5,-8,8]) # Ограниечение от min до max, сначало горизонт последующее вертикаль
plt.plot(x, x2, label ='x2') 
plt.plot(x, x3, label = 'x3')
plt.plot(x, x4, label = 'x4')
plt.title('Граффики f(x)' , fontdict = title_dict)
plt.xlabel('X',fontdict = label_dict)
plt.ylabel('Функция',fontdict = label_dict)
plt.legend(fontsize=8, shadow=True, framealpha=1, facecolor='y', edgecolor='y', title='Легенда')  
plt.show
```

```python
plt.axis([-5,5,-8,8]) # Ограниечение от min до max, сначало горизонт последующее вертикаль
legend_dict = {'family':'Calibri','size':'10'}

plt.plot(x, x2, label='x2') 
plt.plot(x, x3, label ='x3')
plt.plot(x, x4, label ='x4')
plt.title('Граффики f(x)' , fontdict = title_dict)
plt.xlabel('X',fontdict = label_dict)
plt.ylabel('Функция',fontdict = label_dict)
legend = plt.legend(loc = 'lower right', prop=legend_dict, frameon = False ) #Без рамки.
plt.setp(legend.get_texts(), color = 'grey')
plt.show
```

Оформление линий и сетки графика 

```python
plt.axis([-5,5,-8,8])
plt.plot(x, x2, label='x2', color = 'red', linestyle ='dashed')  # Задаем цвет линии 
plt.plot(x, x3, label ='x3', color = 'blue', linestyle ='dotted') # Задаем стиль линии 
plt.plot(x, x4, label ='x4',color = 'green', linestyle ='dashed')
plt.title('Граффики f(x)' , fontdict = title_dict)
plt.xlabel('X',fontdict = label_dict)
plt.ylabel('Функция',fontdict = label_dict)
legend = plt.legend(loc = 'lower right', prop=legend_dict, frameon = False ) #Без рамки.
plt.setp(legend.get_texts(), color = 'grey')
plt.grid(axis = 'y', color = 'black')  #Сетка  по оси Y  черным цветом. 
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5,5,-8,8])
plt.plot(x,x2)
plt.plot(x,x3)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1) #создаем фигуру и аксис
ax.plot(x,x2)
```

### Обьекты figure и axes

```python
fig, ax = plt.subplots(nrows=1,ncols=2) #создаем 2 фигуры и аксис
ax1,ax2 = ax.flatten()    #  ( 1 строка 2 столбца. Достоаем их из массива numpy) 
ax1.plot(x,x2)            # Распремит в одномерный массив
ax2.plot(x,x3)
```

#### Свойства обьекта figure

```python
fig, ax = plt.subplots(nrows=1,ncols=2) #создаем 2 фигуры и аксис
ax1,ax2 = ax.flatten()    #  ( 1 строка 2 столбца. Достоаем их из массива numpy) 
ax1.plot(x,x2)            # Распремит в одномерный массив
ax2.plot(x,x3)
fig.set_size_inches(8,3.5)  # Ширина и высота фигуры
plt.subplots_adjust(wspace=0.7) # Ширина между картинками 
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1) #создаем 2 фигуры и аксис
print(ax.shape)
ax1,ax2 = ax.flatten()    #  ( 1 строка 2 столбца. Достоаем их из массива numpy) 
ax1.plot(x,x2)            # Распремит в одномерный массив
ax2.plot(x,x3)
fig.set_size_inches(3,6)  # Ширина и высота фигуры
plt.subplots_adjust(hspace=0.7) # Ширина между картинками 
```

```python
y=np.sin(x) 
z=np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2, sharey = True) #создаем 4 фигуры и аксис , Оси равны у всех(-100:100)
ax1,ax2,ax3,ax4 = ax.flatten()    #  ( 2 строка 2 столбца. Достоаем их из массива numpy) 
ax1.plot(x,x2)# Распремит в одномерный массив
ax1.set_title('График х2') #Название графика

ax2.plot(x,x3)
ax2.set_title('График х3') #Название графика

ax3.plot(x,y)
ax3.set_title('График sin(х)') #Название графика

ax4.plot(x,z)
ax4.set_title('График cos(х)') #Название графика

fig.set_size_inches(8,6)  # Ширина и высота фигуры
plt.subplots_adjust(wspace=0.3, hspace=0.3) # Ширина между картинками 
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2) #создаем 4 фигуры и аксис
ax1,ax2,ax3,ax4 = ax.flatten()    #  ( 2 строка 2 столбца. Достоаем их из массива numpy) 

ax1.plot(x,x2)# Распремит в одномерный массив
ax1.set_title('График х2') #Название графика
ax1.set_xlim([-3,3]) # Ограничение оси Х

ax2.plot(x,x3)
ax2.set_title('График х3') #Название графика
ax2.set_ylim([-20,20]) # Ограничение оси Y

ax3.plot(x,y)
ax3.set_title('График sin(х)') #Название графика

ax4.plot(x,z)
ax4.set_title('График cos(х)') #Название графика

fig.set_size_inches(8,6)  # Ширина и высота фигуры
plt.subplots_adjust(wspace=0.3, hspace=0.3) # Ширина между картинками 
```
